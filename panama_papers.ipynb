```python
#for col in df.columns:
#    print(col)

df2=df.query("not country_codes.isna()")
df2
```

```python
countries=[]
c=[]
ctr=""
for x in df2.country_codes:
    if (x not in c) :
        c.append(x)


for i in range(0,len(c)-1):
    ctr=''
    for j in c[i]:
        if(j==';') :
            if (ctr not in countries):
                countries.append(ctr)
            ctr=''
        else:
            ctr+=j
    if(ctr not in countries):
        countries.append(ctr)
#print(countries)

print("il y a environ "+str(len(countries))+" pays impliqués")
```
